# Mapping LGD code with Indian States/UTs

Shape data source(state & district): https://www.arcgis.com/home/item.html?id=cba8bddfa0ab43ddb35a7313376f9438

LGD Code Source: https://lgdirectory.gov.in/globalviewstateforcitizen.do?OWASP_CSRFTOKEN=2SQS-KOAV-NBVU-XA8P-Q7D4-AE9E-S7M5-J9BK#

```python
import pandas as pd
import geopandas as gpd

# Reading shape file
states = gpd.read_file('India State and Country shapes(2020)/India country and states/India_State_Boundary.shp')
states
```

```python
# Reading LGD sheet
lgd_df = pd.read_excel('India State and Country shapes(2020)/LGD.xlsx')
lgd_df
```

```python
# Adjusting the state names 
lgd_df.loc[31, 'State Name (In English)'] = states['State_Name'][2]
lgd_df.loc[0, 'State Name (In English)'] = states['State_Name'][0]
lgd_df.loc[12, 'State Name (In English)'] = states['State_Name'][35]
states.loc[13, 'State_Name'] = lgd_df['State Name (In English)'][29]
states.loc[14, 'State_Name'] = lgd_df['State Name (In English)'][6]
states.loc[15, 'State_Name'] = lgd_df['State Name (In English)'][30]
lgd_df
```

```python
# Dropping duplicate - 'puducherry' 
# states.drop(index=17, inplace=True)
# states.reset_index(inplace=True)
# states.drop(columns='index', inplace=True)
states
```

```python
# Mapping right LGD Codes with its shapes
for i in range(len(states.index)):
    state_filt = lgd_df['State Name (In English)'] == states['State_Name'][i]
    states.loc[i, 'LGD Code'] = list(lgd_df.loc[state_filt, 'State LGD Code'])

states
```

```python
# Final Mapping
lgd_mapping = dict(zip(states['LGD Code'], states['geometry']))
lgd_mapping
```
